```python
#1) Fit a Lagrange polynomial to the data in the table above and plot the result.

import numpy as np
from matplotlib import pyplot as plt
from math import sin, pi

x = np.array([0.5, 1, 1.5])
y = np.array([45, 26, 40])


def f(x_new):
    d = x.shape[0] - 1
    s = 0.0
    for i in range(d+1):
        p = y[i]
        for j in range(d+1):
            if i != j:
                p = p * ((x_new - x[j]) / (x[i] - x[j]))
        s = s + p
    return s


tc=np.linspace(0.5, 1.5, 100)  #between 0.5 to 1.5 contain 100 data
ac = np.array([f(i) for i in tc])  #passing 100 data in lagrange function

y_= np.array([f(i) for i in x])


plt.scatter(x, y, color='red')
plt.plot(tc, ac)  
plt.show()

plt.scatter(x, y, color='red')
plt.plot(x, y_)
plt.show()
```
