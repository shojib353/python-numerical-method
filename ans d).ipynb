```python
#1) Scatter plot ùëî(ùë•) for ùë• = ‚àí10, ‚àí9, ‚àí8, ‚Ä¶ 50

import numpy as np
from math import pi
from numpy.random import random
from matplotlib import pyplot as plt
from scipy.linalg import svd
def f(x):
    return np.exp(-x**2)

x = np.arange(-10, 50)
N = x.shape[0]
y= np.array([f(_) for _ in x])  

plt.scatter(x, y,color='red')
plt.show()
```

```python
#2) Fit a 3rd order polynomial regression curve. Use pinv (pseudoinverse)
def basis_expansion(x, degree):
    return np.power(x, np.arange(degree+1))




degree = 3

print(N)
X = np.zeros((N, degree+1))
for i in range(N):
    X[i] = basis_expansion(x[i], degree)
    

psudoInversceX= np.linalg.pinv(X)


w_best = psudoInversceX.dot(y)
print('w_best =', w_best)

y_predicted1 = X.dot(w_best)


plt.scatter(x, y, color='yellow')
plt.plot(x, y_predicted1)
plt.show()
```

```python

#3) Use a regularized least square with ùúÜ = 100. Use SVD



reg = 100
U, d, VT = svd(X,full_matrices=False)

D = np.diag(d / (reg+(d**2)))

w_best = VT.T.dot(D).dot(U.T).dot(y)
print('w_best =', w_best)

y_predicted = X.dot(w_best)


plt.scatter(x, y, color='yellow')
plt.plot(x, y_predicted)
plt.show()

```

```python
#4.Plot the regression curve as a continuous plot.

plt.plot(x,y, label='real function', marker='o')
plt.plot(x,y_predicted, label='regilarize regresstion ', linestyle='--')
plt.legend()
plt.show()

```
